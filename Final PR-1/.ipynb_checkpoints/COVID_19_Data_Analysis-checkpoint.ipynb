This notebook analyzes the spread of COVID-19 across countries and continents, comparing cases, recoveries, deaths and testing data. The dataset used here is a single-day global snapshot (30 June 2024), so instead of a day-by-day time series, the analysis focuses on cross-country and cross-continent comparison, recovery/death rate calculation, and correlation between the different metrics.

**Tools used:** Pandas, NumPy, Matplotlib, Seaborn, Plotly

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

## 1. Loading the Dataset

The `CovidDataAnalyzer` class wraps every step of the workflow, from loading the raw CSV to cleaning it and separating country-level records from continent/global aggregate records.

```python
class CovidDataAnalyzer:
    def __init__(self, filepath):
        self.filepath = filepath
        self.raw_df = None
        self.df = None
        self.country_df = None

    def load_data(self):
        self.raw_df = pd.read_csv(self.filepath)
        self.df = self.raw_df.copy()
        return self.df

    def clean_data(self):
        self.df["continent"] = self.df["continent"].fillna("Unknown")
        self.df["Recovered"] = self.df["Recovered"].fillna(0)
        self.df["Deaths"] = self.df["Deaths"].fillna(0)
        self.df["Tests"] = self.df["Tests"].fillna(0)
        self.df["population"] = self.df["population"].fillna(0)
        return self.df

    def split_country_level(self):
        aggregate_names = ["All", "Africa", "Asia", "Europe", "North-America", "South-America", "Oceania"]
        mask = (~self.df["country"].isin(aggregate_names)) & (self.df["continent"] != "Unknown")
        self.country_df = self.df[mask].copy()
        return self.country_df

    def add_rate_columns(self):
        self.country_df["DeathRate"] = (self.country_df["Deaths"] / self.country_df["Cases"]) * 100
        self.country_df["RecoveryRate"] = (self.country_df["Recovered"] / self.country_df["Cases"]) * 100
        return self.country_df

    def global_summary(self):
        return {
            "Total Countries": self.country_df["country"].nunique(),
            "Total Cases": int(self.country_df["Cases"].sum()),
            "Total Deaths": int(self.country_df["Deaths"].sum()),
            "Total Recovered": int(self.country_df["Recovered"].sum()),
            "Global Death Rate (%)": round(self.country_df["Deaths"].sum() / self.country_df["Cases"].sum() * 100, 2),
        }

    def continent_summary(self):
        return self.country_df.groupby("continent")[["Cases", "Deaths", "Recovered"]].sum().sort_values("Cases", ascending=False)

    def top_n(self, metric, n=10, min_cases=0):
        subset = self.country_df[self.country_df["Cases"] >= min_cases]
        return subset.sort_values(metric, ascending=False).head(n)[["country", metric]]
```

## 2. Loading and Cleaning

```python
analyzer = CovidDataAnalyzer("covid_19.csv")
raw_data = analyzer.load_data()
raw_data.head()
```

```python
print("Missing values before cleaning")
print(raw_data.isnull().sum())
```

```python
clean_data = analyzer.clean_data()
print("Missing values after cleaning")
print(clean_data.isnull().sum())
```

## 3. Separating Country-Level Records from Aggregates

The dataset contains rows for individual countries as well as rows that represent continent totals and a global total (`country == "All"`). These aggregate rows are separated out before doing any country-level comparison, so totals are not double counted.

```python
country_data = analyzer.split_country_level()
country_data = analyzer.add_rate_columns()
print("Country-level records:", country_data.shape[0])
country_data.head()
```

## 4. Global Summary

```python
summary = analyzer.global_summary()
for key, value in summary.items():
    print(f"{key}: {value:,}" if isinstance(value, int) else f"{key}: {value}")
```

## 5. Continent-Wise Analysis

Grouping the country-level data by continent shows how the pandemic's total impact was distributed geographically.

```python
continent_data = analyzer.continent_summary()
continent_data
```

```python
continent_data["Cases"].plot(kind="bar", color="steelblue")
plt.title("Total COVID-19 Cases by Continent")
plt.ylabel("Total Cases")
plt.xlabel("Continent")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

## 6. Top 10 Countries by Confirmed Cases

```python
top_cases = analyzer.top_n("Cases", n=10)

sns.barplot(data=top_cases, x="Cases", y="country", hue="country", palette="Blues_r", legend=False)
plt.title("Top 10 Countries by Total Cases")
plt.xlabel("Total Cases")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

## 7. Top 10 Countries by Deaths

```python
top_deaths = analyzer.top_n("Deaths", n=10)

sns.barplot(data=top_deaths, x="Deaths", y="country", hue="country", palette="Reds_r", legend=False)
plt.title("Top 10 Countries by Total Deaths")
plt.xlabel("Total Deaths")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

## 8. Death Rate and Recovery Rate

Countries with very few reported cases can show misleading rates, so only countries with more than 100,000 cases are considered here.

```python
top_death_rate = analyzer.top_n("DeathRate", n=10, min_cases=100000)
top_recovery_rate = analyzer.top_n("RecoveryRate", n=10, min_cases=100000)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.barplot(data=top_death_rate, x="DeathRate", y="country", hue="country", palette="OrRd_r", legend=False, ax=axes[0])
axes[0].set_title("Highest Death Rate (%)")
axes[0].set_xlabel("Death Rate (%)")

sns.barplot(data=top_recovery_rate, x="RecoveryRate", y="country", hue="country", palette="Greens_r", legend=False, ax=axes[1])
axes[1].set_title("Highest Recovery Rate (%)")
axes[1].set_xlabel("Recovery Rate (%)")

plt.tight_layout()
plt.show()
```

## 9. Relationship Between Cases and Deaths

```python
plt.figure(figsize=(9, 6))
sns.scatterplot(data=country_data, x="Cases", y="Deaths", hue="continent", size="population", sizes=(20, 400), alpha=0.7)
plt.xscale("log")
plt.yscale("log")
plt.title("Cases vs Deaths by Country (log scale)")
plt.xlabel("Total Cases (log)")
plt.ylabel("Total Deaths (log)")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

## 10. Correlation Between Metrics

```python
correlation_columns = ["Cases", "Recovered", "Deaths", "Tests", "population"]
correlation_matrix = country_data[correlation_columns].corr()

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between COVID-19 Metrics")
plt.tight_layout()
plt.show()
```

## 11. Continent-Wise Share of Global Cases

```python
plt.figure(figsize=(7, 7))
plt.pie(continent_data["Cases"], labels=continent_data.index, autopct="%1.1f%%", startangle=90, colors=sns.color_palette("Set2"))
plt.title("Share of Global COVID-19 Cases by Continent")
plt.tight_layout()
plt.show()
```

## 12. Interactive Visualization with Plotly

Interactive charts make it easier to explore individual countries and continents through hover tooltips and zoom.

```python
fig = px.treemap(country_data, path=["continent", "country"], values="Cases", color="DeathRate", color_continuous_scale="Reds", title="COVID-19 Cases by Continent and Country (colored by Death Rate)")
fig.show()
```

```python
fig = px.bar(top_cases, x="country", y="Cases", color="Cases", color_continuous_scale="Blues", title="Top 10 Countries by Total Cases")
fig.show()
```

## 13. Testing vs Confirmed Cases

Comparing the number of tests conducted against confirmed cases gives a sense of how thoroughly each country was screening its population.

```python
testing_data = country_data[country_data["Tests"] > 0]

fig = px.scatter(testing_data, x="Tests", y="Cases", color="continent", hover_name="country", log_x=True, log_y=True, title="Tests Conducted vs Confirmed Cases")
fig.show()
```

## 14. Menu-Driven Visualization Selector

To make the notebook reusable for different chart requests, `display_chart` dispatches to the right plotting logic using `match/case` based on the requested chart type.

```python
class ChartDispatcher:
    def __init__(self, country_df, continent_df):
        self.country_df = country_df
        self.continent_df = continent_df

    def display_chart(self, chart_type):
        match chart_type:
            case "top_cases":
                data = self.country_df.sort_values("Cases", ascending=False).head(10)
                sns.barplot(data=data, x="Cases", y="country", hue="country", palette="Blues_r", legend=False)
                plt.title("Top 10 Countries by Cases")
                plt.tight_layout()
                plt.show()
            case "top_deaths":
                data = self.country_df.sort_values("Deaths", ascending=False).head(10)
                sns.barplot(data=data, x="Deaths", y="country", hue="country", palette="Reds_r", legend=False)
                plt.title("Top 10 Countries by Deaths")
                plt.tight_layout()
                plt.show()
            case "continent_pie":
                plt.figure(figsize=(7, 7))
                plt.pie(self.continent_df["Cases"], labels=self.continent_df.index, autopct="%1.1f%%", colors=sns.color_palette("Set2"))
                plt.title("Cases Share by Continent")
                plt.tight_layout()
                plt.show()
            case "recovery_vs_death":
                fig, ax = plt.subplots(figsize=(8, 6))
                sns.scatterplot(data=self.country_df, x="DeathRate", y="RecoveryRate", hue="continent", ax=ax)
                ax.set_title("Death Rate vs Recovery Rate")
                plt.tight_layout()
                plt.show()
            case _:
                print(f"No chart available for '{chart_type}'")
```

```python
dispatcher = ChartDispatcher(country_data, continent_data)

for chart_type in ["top_cases", "top_deaths", "continent_pie", "recovery_vs_death"]:
    dispatcher.display_chart(chart_type)
```

## 15. Key Insights

- **USA, India and France** recorded the highest total case counts, while **USA, Brazil and India** recorded the highest total deaths.
- **Europe and Asia** together accounted for the largest share of global cases, followed by North America.
- The **global death rate** across all countries in the dataset works out to just under **1%**, though it varies widely by country, from under 0.1% to over 4% in countries like Peru and Egypt.
- Countries such as **South Korea, Singapore and Qatar** show recovery rates above 99%, reflecting strong reporting and treatment outcomes.
- The heatmap shows a strong positive correlation between **Cases and Recovered**, and a moderately strong correlation between **Cases and Deaths**, which is expected since larger outbreaks generally produce more of both outcomes.
- Countries that conducted more tests generally reported more confirmed cases, consistent with the idea that testing capacity strongly influences how much of an outbreak actually gets detected.
